```python
# i_max, j_max = max(g)

# for i in range(i_max):
#     for j in range(j_max):
g = Grid(d)

start = [k for k,v in g.items() if v == '^'][0]
pos = start
dir = (0,-1)
path = []
p1 = 1
while True:
    path.append((pos, dir))
    if add(pos, dir) not in g:
        break
    while g[add(pos, dir)] == '#':
        dir = (-dir[1], dir[0])
    pos = add(pos, dir)
    if g[pos] == '.':
        p1 += 1
        g[pos] = 'X'

# g.print()
p1
```
