```python
from aoc_utils import *

import pandas as pd
import numpy as np

from aocd import get_data, submit
```

```python
year = 2024
day = 18
```

```python
dat = get_data(day=day, year=year, block=True)
```

```python
dat = '''5,4
4,2
4,5
3,0
2,1
6,3
2,4
1,5
0,6
3,3
2,6
5,1
1,2
5,5
2,5
6,5
1,4
0,4
6,4
1,1
6,1
1,0
0,5
1,6
2,0'''
```

```python
fallen = parse(dat, ints)

d = {}
x_max = 71
for ix in range(x_max):
    for iy in range(x_max):
        d[(ix,iy)] = '.'
for (ix,iy) in fallen[:1024]:
    d[(ix,iy)] = '#'
    
g = Grid(d)
```

```python
class MemProblem(GridProblem):
    def actions(self, state):
        loc, time = state
        return [(n, time+1) for n in self.grid.neighbors(loc)]

    def action_cost(self, s, a, s1):
        loc, time = a
        if self.grid[loc] == '.':
            return 1
        else:
            return np.inf
        

    def is_goal(self, state):
        return state[0] == self.goal
        
    def h(self, node):
        loc = node.state[0]
        return taxi_distance(loc, self.goal)

gp = MemProblem(grid=g, initial=((0,0),0), goal=(x_max-1, x_max-1))
p1 = A_star_search(gp)
p1 = p1.path_cost
p1
```

```python
submit(p1, day=day, part='a', year=year)
```

```python
fallen = parse(dat, ints)

f = 1024
d = {}
x_max = 71
for ix in range(x_max):
    for iy in range(x_max):
        d[(ix,iy)] = '.'
for (ix,iy) in fallen[:f]:
    d[(ix,iy)] = '#'
    
g = Grid(d)
class MemProblem(GridProblem):
    def action_cost(self, s, a, s1):
        if self.grid[a] == '.':
            return 1
        else:
            return np.inf

p2 = 0
soln = None
while p2 < np.inf:
    f += 1
    g[fallen[f]] = '#'
    if soln and (fallen[f] not in path_states(soln)):
        continue
    gp = MemProblem(grid=g, initial=(0,0), goal=(x_max-1, x_max-1))
    soln = A_star_search(gp)
    p2 = soln.path_cost
f
```
